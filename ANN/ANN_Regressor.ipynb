### Deep Learning

```python
# Defining ANN model

class ANN(nn.Module):
    def __init__(self):
        super(ANN, self).__init__()

        self.model=nn.Sequential(

            # 1st Hidden layer
            nn.Linear(4,6),
            nn.ReLU(),

            # 2nd Hidden layer
            nn.Linear(6,6),
            nn.ReLU(),

            # output layer

            nn.Linear(6,1),

        )

    def forward(self,x):
        return self.model(x)
```

```python
import torch.optim as optim

model=ANN()

#loss,optimizer

criteria=nn.MSELoss()
optimizer=optim.Adam(model.parameters())
```
