```python
import gymnasium as gym
import numpy as np
import random
```

```python
gamma=0.99
alpha=0.5
epsilon=0.1
episodes=500
```

```python
# Q table => store all q values

Q=np.zeros((48,4))
```

```python
# epsilon policy

def epsilon_greedy(state):
    if random.random() < epsilon:
        return env.action_space.sample()
    else:
        return np.argmax(Q[state])
```

### Q-Learning

```python
for episode in range(episodes):

    render=(episode%50==0)

    if render:
        env=gym.make("CliffWalking-v1",render_mode="human")
        print(f"rendering for episode{episode+1}/500")
    else:
        env=gym.make("CliffWalking-v1")

    done=False

    state,_=env.reset()

    episode_len=0
    rewards=0

    while not done:
        action=epsilon_greedy(state)

        next_state,reward,terminated,truncated,_=env.step(action)

        done=terminated or truncated

        # Q -learning update

        Q[state,action]+=alpha*(reward+gamma*np.max(Q[next_state])-Q[state,action])

        state=next_state

        episode_len+=1
        rewards+=reward
    print(f"for episode {episode+1}/500 rewards is {rewards}")
    env.close()
```

### What did out agent learnt?
